# Exception handling

## try / except

try : 예외 발생 가능 코드

except <Exception type> : 예외 발생시 대응 코드

```python
for i in range(10) :
  try :
    print(10//i)

  except ZeroDivisionError :    # valueError
    print('Not devided by 0')

  except IndexError as e :
    print(e)

  except Exception as e :
    print(e)

```

Built in Exception

- IndexError
- NameError
- ZeroDevisionError
- ValueError
- FileNotFoundError

else

```python
try :
  # 예외 발생
except :
  # 예외 발생시 코드
else :          # finally도 사용 가능
  # 그외 상황이면 실행
```

## assert

- 특정 조건에 만족하지 않을 경우 예외

assert 예외조건

```python
def get_bin_number(decimal_number) :
  assert isinstance(decimal_number, int)
  return bin(decimal_number)

print(get_bin_number(10.0))
```

# log 파일 생성하기



1. 디렉토리가 있는지 확인

2. 파일이 있는지 확인

```python
import os
if not os.path.isdir('log') :
  os.mkdir('log')

if not os.path.exists('log/count_log.txt') :      # 파일이 없으면 생성
  f = open('log/count_log.txt', 'w', encoding = 'utf8')
  f.write('기록 시작 \n')
  f.close()

with open('log/count_log.txt','a', encoding = 'utf8') as f:
  import random, datetime
  for i in range(1,11) :
    stamp = str(datetime.datetime.now())
    value = random.random() * 100000
    log_line = stamp + '\t' + str(value) + '값이 생성되었습니다' + '\n'
    f.write(log_line)
```

pickle 예시

```python
import pickle

class Multiply :
  def __init__(self, multiplier) :
    self.multiplier = multiplier

  def multiply(self, number) :
    return number * self.multiplier

multiply = Multiply(5)
multiply.multiply(10)

f = open('multiply_obj.pickle', 'wb')
pickle.dump(multiply,f)
f.close()
```

```python
f = open('multiply_obj.pickle', 'rb')
multi_pic = pickle.load(f)
multi_pic.multiply(5)
```

##Logging

- 프로그램이 실행되는 동안 정보를 기록화
- 유저의 접근, exception, 특정 함수 사용
- console화면에 출력, 파일에 남기기. DB에 남기기
- 기록된 로그를 분석하여 의미있는 결과 도출
- 실행시점 및 개발시점에 남겨야 하는 기록

print vs logging
- 기록을 print로 남기는 것도 가능
- 로깅의 레벨별로 기록화 필요
- 모듈별 별도의 로깅


```python
import logging

logging.debug('')
logging.info()
logging.warning()
logging.error()
logging.critical()
```

logging level

1. debug : 개발시 처리 기록
2. info : 처리 진행 기록
3. warning : 잘못입력, 의도치 않은 정보
4. error : Exception처리
5. critical : 프로그램 동작이 안됨

```python
import logging

logger = logging.getLogger
logging.basicConfig(level = logging.DEBUG)    # 경고 레벨 설정
logger.setLevel(logging.ERROR)

steam_handler = logging.FileHandler('my.log', mode='a', encoding='utf8')      # my.log라는 파일에 출력할 것임

logger.addHandler(steam_handler)

logger.setLevel(logging.DEBUG)      # 로깅 레벨 설정
logging.debug('')
logging.info()
logging.warning()
logging.error()
logging.critical()
```

argparser

- 콘솔창에서 setting 정보를 저장
- ls --help 쳐보면 됨

- 일반적으로는 argparser사용
- command-line option

```python
from ast import arg
import argparse

parser = argparse.ArgumentParser(description = 'Sum two integers')

parser.add_argument('short name1', "--long_name1", dest = 'A_value', help='A integers', type = int)
parser.add_argument('short name2', "--long_name2", dest = 'B_value', help='B integers', type = int)

args = parser.parse_args()
print(args)
print(args.a)
print(args.b)
print(args.a + args.b)
```
