## 3. row

```python
A = [[1,1,0],
     [1,0,1],
     [1,-1,1]]

X= [1,2,3]
# 선형결합은 다음과 같다.
X = X.T

#전체를 transpose
X.T * A.T

```

# 선형 독립

- 선형 결합에 의한 결과로 해를 판단하는 기준
- 주어진 벡터로 만드는 평행사변형이 unique한 경우

## 선형종속
- 선형 결합에 의해 만들어진 평행사변형이 unique한 것이 아닌 여러 개를 만들 수 있는 경우
- 벡터의 계수가 여러 개인 경우

선형 독립의 실제적 정의

- 각각의 벡터세트를 하나씩 늘렸을 때 기존의 벡터와 겹치지 않아 새로운 span으로 확인 될 수 있는 경우

- 하나의 벡터라도 기존의 벡터와 같은 차원의 방향을 갖는 경우 선형 종속
- 주어진 차원보다 더 많은 벡터가 존재할 경우 그 경우는 무조건 선형 종속

Formal Definition

- Ax = 0일때
- 이를 만족시키는 모든 x가 0이라면 재료벡터에 상관없이 해는 유일하게 0이 된다.
- 이를 tirivial solution이라 한다.
- 만약 x벡터를 만족하는 element중 하나라도 0이 아니면 해가 하나 이상이 되며  선형 종속이 된다.


## 선형 종속은 즉
- 선형 종속 관계에 있는 벡터는 span을 늘리지 않는다
