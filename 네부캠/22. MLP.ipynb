#Neural Networks

- 비선형 변환에 따라오는 행렬 변환 스택 근사 함수 모델

선형모델에서의 netural network

- data D
- model =  y_hat = wx+b
- loss : 1/N sum(y-y_hat)^2

- 선형회귀는 y를 잘 예측하기 위한 기울기와 절편을 찾는 것이 목표이며, 그 결과 나타나는 실제값과 구한 값의 차이의 제곱을 최소화 시키는 것이 목적.

- loss를 w에 대한 편미분 + loss를 b에 대한 편미분 한 결과를 업데이트

# 행렬의 곱셈에 대한 해석
- 두개의 벡터 스페이스에 대한 변환
- 선형성을 갖는 변환이 있을 때 그 변화는 항상 행렬로 표현된다.



image.png

x라는 벡터를 W 기울기와 b를 가진 y로 변환

스택이 진행될 때
- y = W.T.2h * W.T.1x
- 이 경우는 사실상 1단

- W.T.2h = W.T.2p(W.T.1*x)
- p라는 nonlinear-transform(대개 활성화함수)가 필요

## Loss Function
- Regression Task : MSE
  - 무조건 loss를 줄이기 위해 모델을 망치는 경우가 있다.

- Classification Task : CE
  - output이 one hot encoding으로 나오게 됨
  - 내 뉴럴 네트워크의 출력값 중 해당하는 클래스의 값만 증가(불확실성을 감소)
  - 결국 최종값이 아닌 인덱스를 중요시

- Probabilistic Task : MLE
