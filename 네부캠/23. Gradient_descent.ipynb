Generalization

- 일반화 성능
- 학습이 진행되면 training error는 감소하지만 test error는 증가하는 경향성
- training error와 test error의 차이가 generalization gap

Overfitting vs underfitting

- overfitting : 테스트 데이터에 적용 되지 않음
  - 지나친 학습
- underfitting : 학습 데이터에 적용되지 않음
  - 부족한 학습

Cross-validation

- 학습된 모델이 학습되지 않은 데이터를 기준으로 진행하였을 때 얼마나 정확한가 확인
- K-fold

- 일반적으로 cross validation을 진행시켜 hyper parameter를 고정한 후 모든 데이터를 재사용하여 재학습

Bias and Variance

- Bias : 방향성(일치성)
- Variance : 범위(일관성)

Tradeoff

- cost minimizing에 관한 tradeoff
- cost를 줄이는 것은 bias^2 + variance + noise

image.png


Bootstrapping
- 고정된 학습데이터의 일부를 활용하여 여러개의 모델을 만드는 경우
  - 각 모델들의 값이 얼마나 일치하는가


Bagging vs. Boosting
- Bagging(Bootstrapping aggregating)
  - 여러개의 모델을 sub sampling해서 각 output의 평균

- Boosting
  - 모델을 sequntial하게 나누어 각 모델을 specific training하게 만듬
  - 각 모델은 다른 모델이 해결하지 못한 부분을 해결하며 이를 통합
  - 각 모델을 sequential하게 만들어서(weak leaner) 하나의 strong learner를 생성
