Methods

1. Stochastic gradient descent
  - single sample을 통한 업데이트

2. Mini-batch gradient descent
  - 뱃치로 나누어서 업데이트

3. Batch gradient descent
  - 모든 데이터를 사용하여 업데이트

Batch size Matters
- large batch size : sharp minimizer에 도달
- small batch size : flat minimizer에 도달

- small batch size가 더 유리
  - flat minimum은 generalization performance가 높기 때문에 어느정도 오차를 허용

##4. Adagrad

- parameter가 많이 변한 경우는 적게 변화시키고, 적게 변한 경우는 많이 변화
- G에 지금까지 gradient의 변화량을 제곱하여 더한 값을 저장
  - 분모에 저장되기 때문에 값이 클수록 미치는 영향은 작아짐

- G가 무한대로가면 학습이 지연되는 경우가 생김

image.png

##5. Adadelta

- Adagrad의 g가 무한으로 커지는 것을 방지
- 현재 timestamp를 window size에 따른 변화량만을 취급
- parameter가 엄청 클 경우 exponential moving average를 사용하여 방지
  - time window에 대한 평균값을 가지게 하기 때문에 무한히 커지는 것을 방지
