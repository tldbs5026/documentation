##조건문이란?

조건에 따라 특정한 동작을 하게하는 명령어

ex)

- 지하철 앞 차 간격이 10m 이하면 // 속도를 10km이하로 늦춰라
- 사용자가 20세 이하면 // vod를 플레이 하지 마라
- 휴대폰 패턴이 5회 틀리면 // 20초동안 대기상태

조건문은 조건의 기준, 실행하는 명령으로 구성됨

조건의 참, 거짓에 따라 명령이 수행되거나 수행되지 않음

파이썬 --> if,else, elif


```python
# if-else

if <condition> :
  <command-1>
  <command-2>
else :
  <command-3>
  <command-4>
```

```python
print('tell em your age')

my_age = int(input())

if my_age <  30 :
  print('welcome to the club')
else :
  print('you not accepted')
```

- 나이는 현재년도 - 태어난 년도 +1
- 26세 이하 20세 이상 : 대학생
- 20세 미만 17세 이상 : 고등학생

```python
print('당신이 태어난 년도를 입력')

answer = int(input())
year = 2023
if 20 <= year - answer +1 <= 26 :
  print('collegue')
elif 17 <= year - answer + 1< 20 :
  print('high')
elif 14 <= year - answer  + 1< 17 :
  print('middle')
elif 8 <= year - answer +1 < 14 :
  print('elementary')
else :
  print('no student')

```

2. 학점 계산

```python
# A+ : 95, A : 90, B+ : 85, F : 60
score = [38,37,7,16,95,71,63,48,49,66,37]

grade = []
tot = {}
# 조건문을 통해 학점 배분
for i in score :
  if i >=95 :
    grade.append('A+')
  elif i >=90 :
    grade.append('A')
  elif i >= 85 :
    grade.append('B+')
  elif i >= 80 :
    grade.append('B')
  elif i >= 75 :
    grade.append('C+')
  elif i>= 70 :
    grade.append('C')
  elif i>= 65 :
    grade.append('D+')
  elif i>= 60 :
    grade.append('D')
  else :
    grade.append('F')


print(grade)
# 만약 둘을 합치고 싶다면
for i in range(len(score)) :
  tot[score[i]] = grade[i]



print(tot)
```

##반복의 제어

break, continue

```python
# break : 특정 조건에서 반복 종료
for i in range(10) :
  if i == 5 :
    print(i)
    break

print('EOP')

# continue : 특정 조건에서 남은 반복 명령 skip
for i in range(10) :
  if i == 5 : continue    # i가 5면 스킵
  print(i)

print('EOP')
```
