# 시퀀스 데이터

- 소리, 문자열, 주가 등의 순차적으로 들어오는데이터

- 시퀀스 데이터는 독립동등분포(i.i.d)가정을 잘 위배하기 때문에 순서를 바꾸거나 과거 정보에 손실이 발생하면 데이터의 확률분포도 바뀌게 된다.

ex) 개가 사람을 문다 // 사람이 개를 문다



## 시퀀스 데이터 다루기

- 조건부확률

image.png


- 위의 조건부확률은 과거의 모든 정보를 사용하지만 모든 데이터를 분석할 필요는 없다.

- 시퀀스 데이터를 다루기 위해선 길이가 가변적인 데이터를 다룰 수 있는 모델이 필요



- 조건부에 들어가는 데이터의 길이는 가변적이다

잠재 AR 모델
- 바로 이전 정보를 제외한 나머지 정보들을 Ht라는 잠재변수로 인코딩

- 잠재변수 Ht를 신경망으로 통해 반복학습하여 시퀀스 데이터의 패턴을 학습하는 모델이 RNN

# RNN
- MLP와 유사한 모양

- MLP는 과거의 정보를 다룰 수 없다.
  - 입력행렬이 현재인 t번째 데이터만 주어지기 때문

image.png

t번째 잠재변수
- 현재 입력 Xt
- 이전 시점의 잠재변수 Ht-1의 합



## 역전파

- RNN은 이전순서의 잠재변수와 현재의 입력을 활용하여 모델링
- 따라서 RNN의 역전파는 잠재변수의 연결그래프에 따라 순차적으로 계산

image.png

- 이를 Backpropagation Through Time(BPTT)

## BPTT
- BPTT를 통해 RNN의 가중치행렬의 미분을 계산하면 미분의 곱으로 이루어진 항이 계산

image.png

- 박스의 부분이 1보다 커지면 무한히 커지고, 1보다 작으면 무수히 작아짐

- 따라서 시퀀스의 길이를 적절하게 끊는 것이 필요
  - Truncated BPTT라고 부름
  - 미래의 정보를 몇개는 끊고 과거 정보를 몇 개의 블록만 지정하여 계산

## 기울기 소실의 해결책

- 시퀀스 길이가 길어질 경우 BPTT를 통한 계산이 불안정해짐
- Vanilla RNN은 길이가 긴 시퀀스를 처리하는데 문제가 생김
  - 이를 해결하기 위해 LSTM와 GRU
