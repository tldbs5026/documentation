## 5. Label Smoothing
- 데이터 두 개를 섞음
- mix up : 두 이미지의 라벨을 50대 50으로 섞어 decision bound를 부드럽게 하는 것
- cut mix : 두 이미지의 특정 영역을 잘라내어 섞음
