# set

- 값을 순서없이 저장, 중복 제거
- set()

```python
s = set([1,2,3,1,2,3])

s

# 추가
s.add(4)
s
#제거
s.remove(4)
s
# 새로추가
s.update([4,5,6,7])
s
# 새로삭제
s.discard(2)
s

# 모두 삭제
s.clear()
s
```

집합 연산 가능

```python
s1 = set([1,2,3,4,5])
s2 = set([3,4,5,6,7])

u = s1.union(s2)
u

i = s1.intersection(s2)
i

s1 & s2

d = s1.difference(s2)
d

s1 - s2
```

#Dictionary

- {key:value} 형태
- {학번 : 이름, 생년월일, 주소...}

```python
student_info = {2014:'Sungchul', 2015 : 'kkndlas', 2016 : 'asjnkjg'}

for key,value in student_info.items() :     # items() : dict의 모든 데이터 출력
  print('key : ',  key ,' ', 'value : ', value)



for key in student_info.keys() : # key값만 출력
  print(key)

for value in student_info.values() :  # value 출력
  print(value)

# 새로운 값 추가
student_info[2017] = 'qkjwnkf'
```

```python
# queue

from collections import deque

queue = deque()

for i in range(5) :
  queue.append(i)

print(queue)

queue.popleft()   # == pop(0)

queue
```

deque

- rotate, reverse등 linked list의 특성을 지원
- 기존 list 형태의 함수 모두 지원

```python
# queue.rotate()


deq = deque(reversed(queue))
deq

deq.extend([5,6,7])
print(deq)
deq.extendleft([8,9,10])
print(deq)
```

deque는 기존 list보다 효율적인 자료구조

효율적 메모리 구조로 인한 처리 속도 향상

```python
# deque

from collections import deque
import time

start = time.time()
deque_list = deque()

# stack
for i in range(100) :
  for j in range(100) :
    deque_list.append(j)
    deque_list.pop()

print(time.time() - start, 'seconds')
```

```python
#genenral list

import time

start = time.time()

lst = []

for i in range(100) :
  for j in range(100) :
    lst.append(i)
    lst.pop()

print(time.time() - start, 'seconds')
```

# defaultdict

- 딕셔너리 타입의 값에 기본 값을 지정

```python
from collections import defaultdict

d = defaultdict()
print(d)
d['hehe'] = 'boy'

d
```

이를 이용해 단어의 개수를 구할 경우

```python
from collections import defaultdict

text = 'a b d c d a e q s d r tg v s df f g s v'
d = defaultdict(int)    # int를 지정해주면 value의 기본값이 int로 저장됨
t = []
t = text.split(' ')

print(t)

for word in t :
  d[word] +=1

d
```

# Counter

- Sequence type의 data element들의 갯수를 dict로 변환

```python
from collections import Counter

c = Counter(['b','s','b','s','s','s','b'])

c   # 카운팅 변환

c = Counter({'b': 3, 's': 4})
lst = c.elements()
print(list(lst))    # 다시 원상복귀
```

```python
c = Counter(a=4,b=2,c=9,d=4)
d = Counter(a=1,b=2,c=3,d=4)

print(c+d)
print(c-d)

print(c & d)
print( c | d)
```
