# Module 만들기

- 파이썬의 모듈 == py파일
- 같은 폴더에 모듈에 해당하는 .py파일, 사용하는 .py를 저장한 후
- import로 모듈 호출

예시


```python
## fah_converter.py

def convert_c_to_f(celcius_value) :
  return celcius_value * 9.0 / 5 +32


## module_ex.py

import fah_converter

print('Enter az celsius value :')
celsius = float(input())
fahrenheit = fah_converter.convert_c_to_f(celsius)
```
