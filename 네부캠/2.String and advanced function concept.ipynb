# 함수 호출

함수에서 parameter를 전달하는 방식

1. 값에 의한 호출
2. 참조에 의한 호출
3. 객체 참조에 의한 호출

##1. 값에 의한 호출
- 함수에 인자를 넘길 때 값만 넘김.
- 함수 내의 인자 값 변경 시, 호출자에 영향을 주지 않음



##2. 참조에 의한 호출
- 함수에 인자를 넘길 때 메모리 주소를 넘김
- 함수 내 인자 값 변경 시, 호출자의 값도 변환

## 3. 객체 참조에 의한 호출
- 파이썬은 객체의 주소가 함수로 전달됨
- 전달된 객체를 참조하여 변경 시 호출자에 영향을 주지만, 새로운 객체를 만들 경우 호출자에 영향을 주지 않음

```python
def spam(eggs) :
  eggs.append(1)    # lst eggs에 1 추가
  egg = [2,3]      # 새로운 객체인 egg
  return egg

ham = [0]
spam(ham)
print(ham)
```

```python
print('enter the value of x and y')

x,y = map(int, input().split())

print('before swapping')
print(x)
print(y)

# 값만 변화시킴
def swap(x,y) :
  tmp = x
  x = y
  y = tmp
  return x,y
print('after swapping')
swap(x,y)

```

# function Docstring

파이썬 함수에 대한 상세스펙을 사전에 작성
- """로 영역 표시 """
- 설명만 볼 경우 function.'__doc__'

```python
def add_bin(a,b) :
  """
  returns the sum of 2 decimal numbers in binary digits.      # 목적

    parameters :
      a(int) : decimal integer
      b(int) : Another decimal integer

    returns  :
      binary_sum (str)
  """

  binary_sum = bin(a+b)[2:]
  return binary_sum


print(add_bin.__doc__)
```
