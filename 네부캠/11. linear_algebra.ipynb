#선형 방정식

- a1x1+a2x2 + ... + anxn = b 의 형태

```python
a = [a1,a2,a3,a4....,an]

x = [x1,x2,x3...]

a.T + x = b
```

선형시스템은 linear equation의 집합



```python
id = [1,2,3]

weight = [60,65,55]
height = [5.5,5.0,6.0]

smoking = [1,0,1]

life_span = [66,74,78]

linear_system = {66 : 60x1+5.5x2+x3,
                 74 : 65x1+5.0x2,
                 78 : 55x1+6.0x2+x3}

```

```python
A = [[60,5.5,1],
     [65,5.0,0],
     [55,6.0,1]]
x = [x1,x2,x3]
B = [66,74,78]
```

A.T @ x = B의 형태가 된다

## 역행렬

- 정사각 행렬
- A-1 @ A = A @ a-1 == I

A-1 = 1/ad-bc[[d -b],[-c,a]]



역행렬의 조건
- Ax = b  != x = A-1b

- 판별식 det(A) = 0

- 선형변환을 통해 벡터의 길이의 확대 및 축소를 나타냄
