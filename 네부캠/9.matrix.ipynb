#행렬의 이해

- 벡터가 공간에서의 한 점 이라면 행렬은 여러 점들로 구성됨
- 행렬의 행벡터 xi는 i번째 데이터를 의미
- 행렬의 xij는 i번째 데이터의 j번째 변수의 값

## 행렬의 덧셈, 뺄셈, 성분곱, 스칼라곱

- 행렬은 벡터를 원소로 갖는 2차원 배열
- 행렬낄 같은 모양을 가지면 덧셈, 뺄셈
- 성분곱은 벡터와 같다
  - 각 인덱스의 위치끼리 곱셈
- 스칼라곱도 벡터와 차이가 없다.


행렬의 곱셈

- i번째 행벡터와 j번째 열벡터 사이의 내적을 성분으로 가지는 행렬

- Xij * Yjk
  - 행렬곱은 X의 열의 개수와 Y의 행의 개수가 같아야 한다

```python
import numpy as np

x = np.array([[1,-2,3],
              [7,5,0],
              [-2,-1,2]])

y = np.array([[0,1],
              [1,-1],
              [-2,1]])

x @ y
```

- numpy의 np.inner는 i번째 행벡터와 j번째 행벡터 사이의 내적을 성분으로 갖는 행렬을 계산
- 수학의 내적과는 다름
  - 수학에서는 보통 tr(XY.T)를 내적으로 계산
  - 따라서 np.inner를 사용하고 싶으면 Y.T와 해주는 것이?



```python
x = np.array([[1,-2,3],
              [7,5,0],
              [-2,-1,2]])

y = np.array([[0,1],
              [1,-1],
              [-2,1]])

# np.inner(x,y)     # 에러가 뜬다

np.inner(x,y.T) #이렇게 해주면 해결
```

## 행렬은 벡터공간에서 사용되는 연산자로 이해

- 행렬곱을 통해 벡터를 다른 차원의 공간으로 보낼 수 있다
- Z = AX 일때
  - 행렬 A를 이용하여 두 벡터 Z,X를 연결하는 개념

- 이를 이용하여 패턴을 추출, 데이터를 압축할 수 있다
  - 모든 선형변환은 행렬곱으로 계산 가능


역행렬
- 어떤 행렬 A의 연산을 거꾸로 되돌리는 행렬
- 역행렬은 행과 열 숫자가 같고 __행렬식이 0이 아닌 경우__에서 계산 가능
  - 역행렬 연산은 n=m일 때만 가능
  - 이 때 행렬 A의 행렬식이 0이 되면 안됨

- linalg 라이브러리를 이용하여 계산가능

```python
x = np.array([[1,-2,3],
              [7,5,0],
              [-2,-1,2]])

np.linalg.inv(x)
```

```python
x @ np.linalg.inv(x)
```

- 만약 역행렬을 계산할 수 없다면 유사역행렬 또는 무어-펜로즈 역행렬 A+를 이용

- n>=m : A+ = (A.T @ A)inv @ A.T
- n<=m : A+ = A.T @ (A @ A.T)inv
  - n>=m 이면 A+ @ A = I가 성립
  - n<=m 이면 A @ A+ = I가 성립

```python
y = np.array([[0,1],
              [1,-1],
              [-2,1]])
```

```python
# pinv() : 유사 역행렬 계산
np.linalg.pinv(y)
```

```python
np.linalg.pinv(y) @ y
```
