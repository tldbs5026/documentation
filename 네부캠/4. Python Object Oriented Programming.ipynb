##노트북 만들기

- 노트 정리
- content, 내용 제거
- 두 개의 노트북을 합쳐 하나로
- note는 notebook에 삽입
- 최고 300페이지 저장 가능



class scheme

1. method
  - notebook : add_note, remove_note, get_number_of_pages
  - Note : write_content, remove_all

2. variable
  - notebook : title, page_number, notes
  - Note : content

```python
class Note(object) :
  def __init__(self, content = None) :
    self.content = content

  def write_content(self, content) :
    self.content = content

  def remove_all(self) :
    self.content = ''

  def __add__(self, other) :
    return self.content + other.content

  def __str__(self) :
    return self.content
```

```python
class NoteBook(object) :
  def __init__(self, title) :
    self.title = title
    self.page_number = 1
    self.notes = {}

  def add_note(self, note, page = 0) :
    if self.page_number < 300 :
      if page == 0 :
        self.notes[self.page_number] = note
        self.page_number +=1
      else :
        self.notes = {page : note}
        self.page_number +=1
    else :
      print('모든 page가 채워졌습니다')

  def remove_notes(self, page_number) :
    if page_number in self.notes.keys() :
      return self.notes.pop(page_number)

    else :
      print('해당 페이지는 존재하지 않습니다')

  def get_number_of_pages(self) :
    return len(self.notes.keys())
```

## 상속

- 부모클래스로부터 속성과 Method를 물려받은 자식 클래스를 생성

```python
class Person(object) :
  def __init__(self, name, age,gender) :
    self.name = name
    self.age = age
    self.gender = gender


  def __str__(self) :
    return '저는 {0} 입니다. 나이는 {1}입니다.'.format(self.name, self.age)

  def about_me(self) :
    print('제 이름은 {}, 나이는 {}'.format(self.name, self.age))

class Korean(Person):
  pass

first_korean = Korean('Sungchul', 35, 'male')

print(first_korean)
```

```python
class Employee(Person) :    # 부모 클래스 Person에서 상속
  def __init__(self, name, age, gender, salary, hire_date) :
    super().__init__(name, age, gender)   # 부모 객체 사용
    self.salary  = salary     # 내 속성값 1
    self.hire_date  = hire_date # 내 속성값 2

  def do_work(self) :
    print('열심히 일합시다')

  def about_me(self) :  # 부모 클래스 함수 재정의
    super().about_me()  # 부모 클래스의 함수 사용
    print('제 급여는 {0}입니다. 제 입사일은 {1}입니다'.format(self.salary, self.hire_date))
    # 부모의 about_me + 나의 about_me
```

```python
myPerson = Person('jong',32,'male')
myPerson.about_me()
```

```python
my_employeee = Employee('Daeho',32,'male', 30000, '2023-05-01')

my_employeee.about_me()
```

##2. 다형성

- 같은 이름 메소드의 내부 로직을 다르게 작성



```python
class Animal :
  def __init__(self, name) :
    self.name =name

  def talk(self) :
    raise NotImplementedError('Subclass must implement abstract method')



class Cat(Animal) :
  def talk(self) :
    return 'Meow'

class Dog(Animal) :
  def talk(self) :
    return 'bow'


animals = [Cat('Missy'),
           Cat('Mr.Mistoffelees'),
           Dog('Lassie')]

for animal in animals :
  print(animal.name + ' : ' + animal.talk())
```

```python
# dummy class
class Product(object) :
  pass
```

```python
class Inventory(object) :
  def __init__(self) :
    self.__items = []   # Private 변수로 선언 타객체가 접근 못함

  def add_new_item(self, product) :
    if type(product) == Product :
      self.__items.append(product)
      print('new item added')

    else :
      raise ValueError('Invalied Item')


  def get_number_of_items(self) :
    return len(self.__items)
```

```python
my_inv = Inventory()
my_inv.add_new_item(Product())
my_inv.append('abc')
```

다른 사람이 활용하기 위해서

1. Product 객체를 inventory 객체에 추가
2. inventory에는 오직 Product 객체만 들어감
3. Inventory에 Product가 몇 개인지 확인 필요
4. Inventory에 Product items 접근 허용


```python
class Inventory :
  def  __init__(self) :
    self.__items = []

  @property     #property decorator 숨겨진 변수를 반환하게 해줌
  def items(self) :
    return self.__items

my_inventory = Inventory()
my_inventory.add_new_item(Product())
my_inventory.add_new_item(Product())
print(my_inventory.get_number_of_items())

items = my_inventory.items    #Property decorator로 함수를 변수처럼 호출
items.append(Product())

```

## 2. Inner function

- 함수 내 또 다른 함수

```python
def print_msg(msg) :
  def printer() :
    print(msg)

  printer()

print_msg('hello')
```

Closures : inner function을 return 값으로 반환

```python
def print_msg(msg) :
  def printer() :
    print(msg)
  return printer()

another = print_msg('hello')
another
```

```python
def tag_func(tag, text) :
  text = text
  tag = tag

  def inner() :
    return '<{0}>{1}<{0}>'.format(tag,text)

  return inner

h1 = tag_func('title1', 'content1')
h2 = tag_func('title2', 'content2')

print(h1)
print(h2)
```

## decorator

- 복잡한 클로져 함수를 간단하게

```python
def star(func) :
  def inner(*args, **kwargs) :
    print(args[1] * 30)
    func(*args, **kwargs)
    print(args[1] * 30)

  return inner

@star
# star 밑의 printer가 func으로 변경
def printer(msg,mark) :
  print(msg)
printer('hello','ㅆ')
```

```python
def generate_power(exponent) :
  def wrapper(f) :
    def inner(*args) :
      result = f(*args)
      print('>>>>>>>>>', result)
      print(exponent)
      return exponent**result

    return inner
  return wrapper


@generate_power(2)
def raise_two(n) :
  return n**2

raise_two(7)
```
