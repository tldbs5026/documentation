# LSTM
image.png


- xt : input - onehot or encoding
- ht : output(hidden state)
- previous cell state : t까지의 정보를 취합한 summarize
- previous hidden state : t+1의 previous hidden state



##LSTM의 core idea

- forget gate : 어떤 정보를 버릴지 결정
- input gate : 어떤 정보를 cell state에 저장할지 결정
- update cell : update the cell state
- output gate : 업데이트 셀에 저장 된 정보를 output으로 보냄
