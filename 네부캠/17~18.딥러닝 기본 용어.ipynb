# 딥러닝의 중요 키워드
- data : 데이터의 형태와 수
- model : 학습하고자 하는 모델
- loss
- algorithm


## 1. Data

- 데이터는 문제의 유형에 따라 다름

1. classification : 물체 구분
2. semantic segmentation : 각 물체의 경계선 구분
3. detection : bounding box
4. pose estimation : 2d,3d의 스켈레톤 정보
5. visual q&a : 질문에 대한 답

## 2. Model
- Alexnet
- googlenet
- resnet
- densenet
- lstm
- deep autoencoders
- gan

## 3. Loss
- 모델에 대한 학습방법
- 우리가 원하는 것의 근사치
- weight, bias 등의 조정방식
- 문제에 따라 다른 방식을 사용할 필요가 있음

- Regression task : MSE최소화
- Classification task : CrossEntropy
- Probabilistic task : MLE

## 4. Algorithm
- SGD : 1차 미분 정보 활용
- optimizer
