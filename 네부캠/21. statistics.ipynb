# 모수

- 통계적 모델링능 적절한 가정 위에서 확률분포를 추정
- 유한한 개수의 데이터만 관찰해서 모집단의 분포를 정확하게 알아낼 수가 없기 때문에 근사적으로 확률분포를 추정할 수 밖에 없다.
  - 분포를 정확하게 맞추는 것보다는 불확실성을 고려해서 위험을 최소화

- 모수적(marametric) 방법론 : 데이터가 특정 확률분포를 따른다고 선험적(a priori)으로 가정한 후 그 분포를 결정하는 모수(parameter)를 추정하는 방법

- 비모수(non-parametric) 방법론 : 특정 확률분포를 가정하기 않고 데이터에 따라 모델의 구조, 모수의 개수를 변경

## 확률분포를 가정하는 방법

- 히스토그램을 통해 모양을 관찰

1. 데이터가 2개의 값(0,1) : 베르누이 분포
2. n개의 이산적인 값 : 카테고리분포
3. [0,1]사이 : 베타분포
4. 데이터가 0이상 : 감마,로그정규분포
5. 데이터가 R전체에서 값 : 정규분포, 라플라스분포

- 기계적으로 확률분포를 가정해서는 안 되며, 데이터를 생성하는 원리를 먼저 고려하는 것이 원칙
  - 각 분포마다 검정하는 방법이 있기 때문에 모수를 추정한 후에는 검정을 진행

# 최대가능도(MLE) 추정법

- 확률분포마다 사용하는 모수가 다르므로 적절한 통계량이 달라짐
- 이론적으로 가장 가능성이 높은 모수를 추정하는 방법 중 하나

image.png


가능도 함수는 모두 theta를 따르는 분포가 x를 관찰할 가능성

- 확률함수 : theta가 주어질 때 x를 변형
- 가능도 함수 : 주어진 데이터 x에 대해 모수 theta를 변수로 둔 함수
  - 데이터가 주어진 상황에서 theta를 변형하는 함수

- 데이터 집합 x가 독립적으로 추출되었을 경우 로그가능도를 최적화

image.png

##로그가능도를 사용하는 이유

- 모수 theta는 가능도를 최적화하는 MLE
- 데이터의 숫자가 커질수록 유리
- 데이터가 독립일 경우, 곱셈을 덧셈으로 교환가능
- 경사하강법으로 가능도를 최적화할 때 미분을 사용하는데, 이 때의 연산량은 O(n^2)에서 O(n)
- 대개의 손실함수의 경우 경사하강법을 사용함으로 음의 로그가능도를 최적화
