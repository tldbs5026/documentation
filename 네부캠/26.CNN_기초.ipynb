## 다양한 차원에서의 Convolution

1. 1D : 하나의 좌표계
  - 음성, 텍스트
2. 2D : 두개의 좌표계
  - 흑백영상
3. 3D : 세개의 좌표계

- 차원이 높아져도 커널의 값은 달라지지 않는다.

## 2차원 convolution
- 커널을 입력벡터 상에서 움직여가며 선형모델과 합성함수가 적용되는 구조

image.png

커널의 영역에 해당하는 입력벡터와의 element-wise 곱셈

- 입력 크기를 (H,W), 커널의 크기를 (KH,KW), 출력 크기를 (OH,OW)일 떄
  - OH = H-KH + 1
  - Ow = W-KW + 1

- 채널이 여러개인 2차원 입력의 경우 2차원 convolution을 채널 개수만큼 적용
- 3차원 부터는 행렬이 아닌 tensor

- 채널이 여러개인 경우 커널의 채널 수와 입력의 채널수가 같아야 한다.

- 채널이 여러개인 2차원의 경우 2차원 convolution을 채널 개수만큼 적용
