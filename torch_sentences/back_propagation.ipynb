```python
import torch
```

변수 선언(+데이터 입력)

```python
x=torch.ones(2,2, requires_grad = True)
#x에 대해 미분이 가능하도록 x에 대한 모든 연산을 출력가능

print(x)
```

모델 내 연산 예측값 산출

```python
y=x+1
print(y)
```

손실함수 계산

```python
z=2*y**2

res = z.mean()
```

손실 산출

```python
print(res)
```

최적화는 손실함수를 바탕으로 미분

```python
# d(res)/dx_i = x_i +1
# res = z_1+...+z_4 / 4
# z_i= 2y_1 **2
#이 모든 과정들이 : 
res.backward()      #이거임
```

```python
print(x.grad)
```
